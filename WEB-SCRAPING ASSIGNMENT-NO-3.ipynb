```python
import time
import pandas as pd
from selenium import webdriver 
from selenium.webdriver.common.action_chains import ActionChains 
from selenium.webdriver.common.by import By
from bs4 import *
import requests
```

```python
driver=webdriver.Chrome("chromedriver.exe")
driver.get("https://www.amazon.in/")
```

```python
search = driver.find_element_by_id("twotabsearchtextbox")
search.send_keys(input())
```

```python
search_btn=driver.find_element_by_id("nav-search-submit-button")
search_btn.click()
```

```python
data = []    
page_count = 1
columns = ["Brand Name","Name of the Product","Price"]

for i in range(3):
    time.sleep(3)
    products = driver.find_elements(By.XPATH,"//div[@class='a-section a-spacing-medium a-text-center']//div[@class='a-section a-spacing-none']")
    print(f"Products in page {page_count}: ",len(products))
    
    for product in products:
        
        try:
            b_name=product.find_element(By.XPATH,".//h5[@class='s-line-clamp-1']/span").text
            data_row = product.text.split("\n")
        except:
            b_name="-"
        try:
            name=product.find_element(By.XPATH,".//span[@class='a-size-base-plus a-color-base a-text-normal']").text
        except:
            name="-"
        try:
            price=product.find_element(By.XPATH,".//span[@class='a-price-whole']").text
        except:
            price="-"
                    
        data.append([b_name,name,price])
        
    page_count += 1
    driver.find_element_by_xpath("//li[@class='a-last']").click()
    time.sleep(2)
print("Done Scraping!!!")
```

```python
df=pd.DataFrame(data,columns=columns)
df
```

```python
rating=[]
n_rating=[]
av=[]
r=[]
e=[]
for i in page_url: 
        driver.get(i)
        time.sleep(3)
        try:
            ratings = driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']")
            rating.append(ratings.text)
        except:
            ratings=driver.find_element_by_xpath("//h3[@class='a-spacing-small']")
            rating.append(ratings.text)
        
        try:
            n_ratings = driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
            n_rating.append(n_ratings.text)
        except:
            n_ratings="-"
            n_rating.append(n_ratings)
        
        try:
            avs= driver.find_element_by_xpath("//div[@id='availability']")
            av.append(avs.text)
        except:
            avs="-"
            av.append(avs)
        
        try:
            es= driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']")
            e.append(es.text)
        except:
            es="-"
            e.append(es)
```

```python
df["product URL"]=page_url
df["ratings"]=rating
df["No.of.ratings"]=n_rating
df["availability"]=av
df["expected delivery"]=e
```

```python
driver=webdriver.Chrome("chromedriver.exe")
driver.get("https://www.forbes.com/billionaires/")
```

```python
Rank=[]
Person_Name=[]
total_net_worth=[]
Age=[]
country_of_citizenship=[]
Source=[]
industry=[]

while(True):
    
    rank= driver.find_elements_by_xpath("//div[@class='rank']")
    for i in rank:
        Rank.append(i.text)
    time.sleep(1)
    
    name= driver.find_elements_by_xpath("//div[@class='personName']//div")
    for i in name:
        Person_Name.append(i.text)
    time.sleep(1)
    
    net= driver.find_elements_by_xpath("//div[@class='netWorth']")
    for i in net:
        total_net_worth.append(i.text)
    time.sleep(1)
    
    ages= driver.find_elements_by_xpath("//div[@class='age']")
    for i in ages:
        Age.append(i.text)
    time.sleep(1)
    
    country= driver.find_elements_by_xpath("//div[@class='countryOfCitizenship']")
    for i in country:
        country_of_citizenship.append(i.text)
    time.sleep(1)
    
    source= driver.find_elements_by_xpath("//div[@class='source']")
    for i in source:
        Source.append(i.text)
    time.sleep(1)
    
    indus= driver.find_elements_by_xpath("//div[@class='category']")
    for i in indus:
        industry.append(i.text)
    time.sleep(1)
    
     
    try:
        next_button = driver.find_element_by_xpath("//button[@class='pagination-btn pagination-btn--next ']")
        next_button.click()
    except:
        break  
```

```python
df=pd.DataFrame()
df["RANK"]=Rank
df["NAME"]=Person_Name
df["NET_WORTH"]=total_net_worth
df["AGE"]=Age
df["COUNTRY_OF_CITIZENSHIP"]=country_of_citizenship
df["SOURCE"]=Source
df["INDUSTRY"]=industry
```

```python
driver=webdriver.Chrome("chromedriver.exe")
driver.get("https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2021-10-25&to=2021-10-28&guests=2&page=1")
```

```python
page_url=[]
data = []  
page_count = 1
columns = ["hostel name","distance from city centre"]

for i in range(3):
    time.sleep(3)
    hostels = driver.find_elements(By.XPATH,"//div[@class='property']//div[@class='title-row']")
    print(f"hostels in page {page_count}: ",len(hostels))
    
    for hostel in hostels:
        
        try:
            h_name=hostel.find_element(By.XPATH,".//h2[@class='title title-6']").text
            data_row = hostel.text.split("\n")
        except:
            h_name="-"
        time.sleep(2)
        
        try:
            distance=hostel.find_element(By.XPATH,".//div[@class='subtitle body-3']").text
        except:
            distance="-"
        time.sleep(2)
        
        try:
            url=hostel.find_element(By.XPATH,".//h2[@class='title title-6']/a").get_attribute('href')
        except:
            url="-"
        
        data.append([h_name,distance])
        page_url.append(url)
    page_count += 1
    try:
        driver.find_element_by_xpath("//div[@class='pagination-item pagination-next']").click()
        time.sleep(2)
    except:
        driver.find_element_by_xpath("//div[@class='pagination-item pagination-first']").click()
        time.sleep(2)
        
print("Done Scraping!!!")
```
