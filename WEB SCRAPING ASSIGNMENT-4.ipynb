```python
driver=webdriver.Chrome("chromedriver.exe")
driver.get("https://www.imdb.com/list/ls095964455/")
```

```python
data=[]
page_url=[]
columns=["name","year span","genre","run time"]
series=driver.find_elements(By.XPATH,"//div[@class='lister-item mode-detail']//div[@class='lister-item-content']")
for serie in series:
    try:
        name=serie.find_element(By.XPATH,".//h3[@class='lister-item-header']/a").text
        data_row = serie.text.split("\n")
    except:
        name="-"
    try:
        y=serie.find_element(By.XPATH,".//span[@class='lister-item-year text-muted unbold']").text 
    except:
        y="-"
    try:
        g=serie.find_element(By.XPATH,".//span[@class='genre']").text 
    except:
        g="-"
    try:
        r=serie.find_element(By.XPATH,".//span[@class='runtime']").text 
    except:
        r="-"
    try:
        url=serie.find_element(By.XPATH,".//h3[@class='lister-item-header']/a").get_attribute("href")
    except:
        url="-"
    
    page_url.append(url)
    data.append([name,y,g,r])
```

```python
rate=[]
vote=[]
for i in page_url:
    driver.get(i)
    time.sleep(3)
    
    try:
        rates = driver.find_element_by_xpath("/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[1]/div[2]/div/div[1]/a/div/div/div[2]/div[1]/span[1]")
        rate.append(rates.text)
    except:
        rates="-"
        rate.append(rates)
    
    try:
        votes = driver.find_element_by_xpath("/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[1]/div[2]/div/div[1]/a/div/div/div[2]/div[3]")
        vote.append(votes.text)
    except:
        votes="-"
        vote.append(votes)
```

```python
df=pd.DataFrame(data,columns=columns)
df["rating"]=rate
df["votes"]=vote
df
```

```python
driver=webdriver.Chrome("chromedriver.exe")
driver.get("https://archive.ics.uci.edu/")
```

```python
recruit=driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td/span/b/a").get_attribute("href")
driver.get(recruit)
```

```python
name=[]
d_type=[]
task=[]
a_type=[]
n_instance=[]
n_attribute=[]
y=[]

names = driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr/td[1]')
for i in names:
    name.append(i.text)

d_types = driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr/td[2]')
for i in d_types:
    d_type.append(i.text)
    
tasks= driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr/td[3]')
for i in tasks:
    task.append(i.text)
    
a_types= driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr/td[4]')
for i in a_types:
    a_type.append(i.text)
    
n_instances= driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr/td[5]')
for i in n_instances:
    n_instance.append(i.text) 

n_attributes = driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr/td[6]')
for i in n_attributes:
    n_attribute.append(i.text)
    
ys = driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr/td[7]')
for i in ys:
    y.append(i.text)
```

```python
df=pd.DataFrame()
df["dataset_name"]=name[1:]
df["data_type"]=d_type[1:]
df["task"]=task[1:]
df["attribute_type"]=a_type[1:]
df["no_of_instances"]=n_instance[1:]
df["no_of_attribute"]=n_attribute[1:]
df["year"]=y[1:]
df
```
