```python
page1=requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
page2=requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt")
print(page1)
print(page2)
```

```python
soup1=BeautifulSoup(page1.content)
soup2=BeautifulSoup(page2.content)
```

```python
movie1=soup1.find_all("h3",class_="lister-item-header")
movie1
```

```python
name=[]
for i in movie1:
    for j in i.find_all("a",href=True):
        name.append(j.text.strip().replace("\n",""))
print(name)
year=[]
for i in movie1:
    for j in i.find_all("span",class_="lister-item-year text-muted unbold"):
        year.append(j.text.strip().replace("\n",""))
print(year)        
```

```python
rating1=soup1.find_all("div",class_="inline-block ratings-imdb-rating")
rating1
```

```python
imdb_rating=[]
for i in rating1:
    imdb_rating.append(i.text.strip())
imdb_rating
```

```python
import pandas as pd
top_rated1=pd.DataFrame()
top_rated1["movies"]=name
top_rated1["imdb_rating"]=imdb_rating
top_rated1["year_of_release"]=year
top_rated1
```

```python
movie2=soup2.find_all("h3",class_="lister-item-header")
movie2
```

```python
name=[]
for i in movie2:
    for j in i.find_all("a",href=True):
        name.append(j.text.strip().replace("\n",""))
print(name)
year=[]
for i in movie2:
    for j in i.find_all("span",class_="lister-item-year text-muted unbold"):
        year.append(j.text.strip().replace("\n",""))
print(year)        
```

```python
rating2=soup2.find_all("div",class_="inline-block ratings-imdb-rating")
rating2
```

```python
imdb_rating=[]
for i in rating2:
    imdb_rating.append(i.text.strip())
imdb_rating
```

```python
top_rated2=pd.DataFrame()
top_rated2["movies"]=name
top_rated2["imdb_rating"]=imdb_rating
top_rated2["year_of_release"]=year
top_rated2
```

```python
top_rated_movies=pd.concat([top_rated1,top_rated2],ignore_index=True)
top_rated_movies
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
page
```

```python
soup=BeautifulSoup(page.text)
soup
```

```python
table=soup.find("div",class_="rankings-block__container full rankings-table")
table
```

```python
header=[]
for i in table.find_all("th"):
    header.append(i.text.strip().replace("\n",""))
header
df=pd.DataFrame(columns=header)
df
```

```python
for row in table.find_all("tr")[1:]:
    data=row.find_all("td")
    row_data=[td.text.strip().replace("\n","-") for td in data]
    length=len(df)
    df.loc[length]=row_data
df
```

```python
df=df.head(10)
df
```
